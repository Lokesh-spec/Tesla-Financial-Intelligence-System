```python
import os
from dotenv import load_dotenv
load_dotenv()

if os.environ['OPENAI_API_KEY']:
    print("API Key is set.")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model='gpt-5-nano', temperature=0)
```

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(model='text-embedding-3-small', dimensions=1024)
```

```python
from langchain_chroma import Chroma
```

```python
vectorstore_persist = Chroma(
    persist_directory='/Users/lokeshkv/data-engineering/Tesla_Financial_Document_Q_and_A_System_using_RAG/data/vector_db_path',
    embedding_function=embedding_model
)
```

```python
# Pass the query string as the first parameter or specify it by name
results = vectorstore_persist.similarity_search(
    query="What was Tesla's total revenue in 2023?",
    k=5,
)
```

```python
for r in results:
    print(r.metadata)
    print(r.page_content[:300])
    print("-----")
```
