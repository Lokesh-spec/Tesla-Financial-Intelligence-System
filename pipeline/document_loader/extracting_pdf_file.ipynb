```python
import os
import yaml
from pathlib import Path
from langchain_community.document_loaders import PyPDFLoader

from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
config_path = Path("/Users/lokeshkv/data-engineering/Tesla_Financial_Document_Q_and_A_System_using_RAG/config/parameters.yaml")

if not config_path.exists():
    print(f"Looking for file at: {config_path.absolute()}")
else:
    with open(config_path, "r") as f:
        config = yaml.safe_load(f)
```

```python
pdf_path = config['pipeline']['raw_pdf_path']
c_size = config['ingestion']['chunk_size']
c_overlap = config['ingestion']['chunk_overlap']
extracted_text_path = config['pipeline']['extracted_text_path']
```

```python
pdf_files = list(Path(pdf_path).glob("*.pdf"))

for pdf_file in pdf_files:
    loader = PyPDFLoader(
        file_path=str(pdf_file),
        mode="page"
    )
    pdf_content = loader.load()

    output_file = os.path.join(extracted_text_path, f"{pdf_file.stem}.txt")
    
    os.makedirs(os.path.dirname(output_file), exist_ok=True)

    with open(output_file, "w", encoding="utf-8") as f:
        for page in pdf_content:
            f.write(page.page_content)
            f.write("\n\n")
    break
```
